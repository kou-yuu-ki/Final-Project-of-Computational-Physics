#### The Effect of Various Imaginary External Bx on its Magnetic Properties

In the second part of project, we move on to discuss how the magnetic properties of the system change with the variation of imaginary external magnetic field in x direction, instead of the real external one in z direction in the previous part.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.sparse.linalg import *
import scipy.sparse as sparse
```

As what we've done in the previous part, we create $S^x$, $S^y$, $S^z$ as Pauli matrices multiplied with $\frac{1}{2}$, and define the Hamiltonians $H_x$, $H_{xx}$, $H_z$ as functions of $N$, the number of the spin particles, below.
$$ H_x(N) := \sum_{i}^{N} S^x_{i} $$
$$ H_{xx}(N) := \sum_{i, i+1}^{N} S^x_{i} S^x_{i+1} $$
$$ H_z(N) := \sum_{i}^{N} S^z_{i} $$
And according to the definition of Heisenberg spin Hamiltonian with imaginary external field:

$$H = -hH_z-JH_{xx}-ih_xH_{x}$$

where $h_x$ represents the imaginary external magnetic field in x-direction, the real external magnetic field in z-direction $h$ is fixed here and $J$ is a ratio constant.

We'll later find out the eigenvalues and eigenvectors of Heisenberg spin Hamiltonian to illustrate the physical picture of this spin system.

```python
Sx = np.array([[0,0.5],[0.5,0]])
Sy = np.array([[0,-0.5j],[0.5j,0]])
Sz = np.array([[0.5,0],[0,-0.5]])
I = np.array([[1,0],[0,1]])
sparse.csr_matrix(Sx)
sparse.csr_matrix(Sy)
sparse.csr_matrix(Sz)
sparse.csr_matrix(I)

# Useful algorithm: rotation
def Hx(N):
    M = [Sx]+ [I]*(N-1) + [Sx]+ [I]*(N-1)  #append: For example, [Sx, I, I, I] + [Sx, I, I, I] = [Sx, I, I, I, Sx, I, I, I]
    ans = []
    for i in range(N):
        tmp = np.array([1,])
        for j in range(N):
            tmp = sparse.kron(tmp, M[i + j],format='csr')  #M[i + j]: swift by one index with the change in i, and sweep with the change in j
        ans.append(tmp)
    return sum(ans)

def Hxx(N):
    M = [Sx]*2+ [I]*(N-2) + [Sx]*2+ [I]*(N-2)  #append: For example, [Sx, Sx, I, I] + [Sx, Sx, I, I] = [Sx, Sx, I, I, Sx, Sx, I, I]
    ans = []
    for i in range(N):
        tmp = np.array([1,])
        for j in range(N):
            tmp = sparse.kron(tmp, M[i + j],format='csr')  #M[i + j]: swift by one index with the change in i, and sweep with the change in j
        ans.append(tmp)
    return sum(ans)

def Hz(N):
    M = [Sz]+ [I]*(N-1) + [Sz]+ [I]*(N-1)  #append: For example, [Sz, I, I, I] + [Sz, I, I, I] = [Sz, I, I, I, Sz, I, I, I]
    ans = []
    for i in range(N):
        tmp = np.array([1,])
        for j in range(N):
            tmp = sparse.kron(tmp, M[i + j],format='csr')  #M[i + j]: swift by one index with the change in i, and sweep with the change in j
        ans.append(tmp)
    return sum(ans)
```

Then, the Hamiltonians for $N$ = 4, 6, 8, 10, 12, 14 are calculated below, and 2 figures are plotted:

#### Figure.1: the relation betweeen $E_1-E_0$ (energy gap) and $h_x$

#### Figure.2: the relation betweeen $1/(E_1-E_0)N$ and $h_x$

where $E_0(=E)$ is defined as the first eigenvalue of $H$, and $E_1$ is defined as the second eigenvalue of $H$.

The energy gap is then defined as $E_1-E_0$.

The meaning of $1/(E_1-E_0)N$ is $\frac{\xi}{N}$, where $\xi$ is defined as $1/(E_1-E_0)$, called "coherence length".

```python
fig, ax = plt.subplots(ncols=2, nrows=1, figsize=(10*2,8))
ax = ax.flatten()

for N in [4, 6, 8, 10, 12, 14]:
    hz = 1
    hx = np.linspace(0.0,0.3)
    J = 1
    lam = J/2
    E0 = np.zeros_like(hx, dtype = complex)
    E1 = np.zeros_like(hx, dtype = complex) 
    mx = np.zeros_like(hx, dtype = complex)
    mz = np.zeros_like(hx, dtype = complex) 
    sparse.csr_matrix(E0)
    sparse.csr_matrix(E1)
    sparse.csr_matrix(mx)
    sparse.csr_matrix(mz)

    for i in range(hx.shape[0]):
        H = -J*Hxx(N)-hz*Hz(N)-hx[i]*Hx(N)*1j
        w, v = eigs(H, 2, which='SR')
        w = np.sort_complex(w)
        E0[i] = np.real(w[0])
        E1[i] = np.real(w[1])
        mx[i] = np.real(v[:,1]@Hx(N)@v[:,0]/N)
        mz[i] = np.real(v[:,0]@Hz(N)@v[:,0]/N)  
    
    # gap
    ax[0].plot(hx, E1-E0, '*-',label='N={}'.format(N))
    ax[0].legend(fontsize=16)
    ax[0].set_title(r"Figure.3: the relation betweeen $E_1-E_0$ (energy gap) and $hx$",fontsize=20)
    ax[0].set_xlabel(r'$h_x$',fontsize=24)
    ax[0].set_ylabel(r'$E_1-E_0$',fontsize=24)
    ax[0].grid()
    
    # xi/L
    ax[1].plot(hx, 1/((E1-E0)*N), '*-', label='N={}'.format(N))
    ax[1].legend(fontsize=16)
    ax[1].set_title(r"Figure.6: the relation betweeen $\xi/N$ and $hx$",fontsize=20)
    ax[1].set_xlabel(r'$h_x$',fontsize=24)
    ax[1].set_ylabel(r'$1/(E_1-E_0)N$',fontsize=24)
    ax[1].semilogy()
    ax[1].grid()   
```

Now, we are interested in the intersections of the functions of $N$ and $N+2$ in Figure.2, so we zoom in the part where the intersections occurs. But as for the data that are freakingly high, we don't want to take them since they are random. Thus, we don't take the values larger than $10^4$ here, and to prevent from determining wrong intersections on the right hand side of the figure, I designed an algorithm that allows the list to gain where the intersections occur only for once. If there are other intersections after that, then they must be from the random data so we won't take them.

```python
fig, ax = plt.subplots(ncols=1, nrows=1, figsize=(10,8))
save = np.zeros([100, hx.shape[0]+1])
sparse.csr_matrix(save)
for N in [4, 6, 8, 10, 12, 14]:
    hx = np.linspace(0.155,0.164)
    E0 = np.zeros_like(hx, dtype = complex)
    E1 = np.zeros_like(hx, dtype = complex)
    xi = np.zeros_like(hx, dtype = complex)
    sparse.csr_matrix(E0)
    sparse.csr_matrix(E1)
    sparse.csr_matrix(xi)

    for i in range(hx.shape[0]):
        H = -J*Hxx(N)-hz*Hz(N)-hx[i]*Hx(N)*1j
        w, _ = eigs(H, 2, which='SR')
        w = np.sort_complex(w)
        E0[i] = np.real(w[0])
        E1[i] = np.real(w[1])
        if np.log(1/((E1[i]-E0[i])*N)) < 4:
            xi[i] = 1/((E1[i]-E0[i])*N)
            save[N][i] =  1/((E1[i]-E0[i])*N)
# xi/L
    ax.plot(hx, xi, '*-', label='N = {}'.format(N))
    ax.legend(fontsize=16)
    ax.set_xlabel(r'$h_x$',fontsize=24)
    ax.set_ylabel(r'$Re(1/(E_1-E_0)N)$',fontsize=24)
    ax.semilogy()
    ax.grid()
```

After zooming in, we can see the intersections clearly in the figure above right now. Then here, we use the method of interpolation and root location theorem to make the discrete data linearly continuous near the intersections happen. The place where the intersections happen is defined as $h_x^*$ here. After plotting $h_x^*$ with respect to $1/N$, we got the figure below.

```python
ig, ax = plt.subplots(ncols=1, nrows=1, figsize=(10,8))
x = np.zeros(100)
for N in [4, 6, 8, 10, 12]:
    for i in range(hx.shape[0]):
        if ((save[N][i]-save[N+2][i])*(save[N][i+1]-save[N+2][i+1]) < 0) & (x[int((N-4)/2)] == 0):
            m1 = (save[N][i]-save[N][i+1])/(hx[i]-hx[i+1])
            m2 = (save[N+2][i]-save[N+2][i+1])/(hx[i]-hx[i+1])
            x[int((N-4)/2)] = (save[N+2][i]-save[N][i])/(m1 - m2) + hx[i]

Ns = [1/4, 1/6, 1/8, 1/10, 1/12]
# h*
ax.plot(Ns, x[:len(Ns)], '*-')
ax.set_xlabel(r'$1/N$',fontsize=24)
ax.set_ylabel(r'$h_x^*$',fontsize=24)
ax.grid()
```

In order to find out the critical $h_x$, ${h_{x}}_c$, which is the value of $h_x^*$ when $N$ approaches infinity, or when $1/N$ approaches $0$. Thus, We fit the data in the figure above with an exponential function $ax^{-b}+c$ and determine the y-intercept $c$ as ${h_{x}}_c$.

```python
from scipy.optimize import curve_fit
def func(x, a, b, c):
    return a*(x**(-b))+ c
xdata = np.array([4, 6, 8, 10, 12])
ydata = x[:len(xdata)]
popt, pcov = curve_fit(func, xdata, ydata)
print("[a, b, c] = ", popt)
critical_hx = popt[2]
print("The critical hx is: ",critical_hx)
```

From the fitting, we obtained the value of ${h_{x}}_c$, which is approximately $0.156$. In the end, to check the quality of fitting, we plotted the original simulation data and the function of the result of fitting together.

```python
ig, ax = plt.subplots(ncols=1, nrows=1, figsize=(10,8))
for N in [4, 6, 8, 10, 12]:
    for i in range(hx.shape[0]):
        if ((save[N][i]-save[N+2][i])*(save[N][i+1]-save[N+2][i+1]) < 0) & (x[int((N-4)/2)] == 0):
            m1 = (save[N][i]-save[N][i+1])/(hx[i]-hx[i+1])
            m2 = (save[N+2][i]-save[N+2][i+1])/(hx[i]-hx[i+1])
            x[int((N-4)/2)] = (save[N+2][i]-save[N][i])/(m1 - m2) + hx[i]

test_recipical_N = np.linspace(0, 0.3)

# h* fitting
ax.plot(1/np.array(xdata), x[:len(xdata)], '*', label='data points')
ax.plot(test_recipical_N, func(1/test_recipical_N, *popt), '-', label='result of fitting')
ax.legend(fontsize=16)
ax.set_xlabel(r'$1/N$',fontsize=24)
ax.set_ylabel(r'$h_x^*$',fontsize=24)
ax.grid() 
```

Yes, from the figure above, we can see that the quality of fitting looks very good. And this is the end of the second part of the simulation project.
